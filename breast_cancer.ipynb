```python
df_mean=df.iloc[:,:11]
```

### drop columns with a very high correlation between another columns

```python
df_mean.drop(['perimeter_mean','area_mean','concave points_mean'],axis=1,inplace=True)
```

### a small function to make two hists

```python
def plott(co):
    plt.subplot(1,2,1)
    sns.distplot(df_mean[co][df_mean['diagnosis']=='M']).set_title('malignant')
    plt.subplot(1,2,2)
    sns.distplot(df_mean[co][df_mean['diagnosis']=='B']).set_title('benign')
    plt.show()
```

```python
plott('radius_mean')
```

### in the second two cells we found that the malignants mean is about 17.5 and the 75% of benign is 13.3
### and that's a big difference to be considered
